```python
from google.colab import files

import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
from PIL import Image
from google.colab.patches import cv2_imshow
```

```python
def analisar_plantacao(caminho_imagem):

  img_pil = Image.open(caminho_imagem)


  img_rgb = np.array(img_pil.convert('RGB'))

  img = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2BGR)

  img_hsv = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)

  verde_baixo = np.array([40, 40, 80])
  verde_alto = np.array([80, 255, 255])

  mascara = cv2.inRange(img_hsv, verde_baixo, verde_alto)
  destaque = cv2.bitwise_and(img, img, mask=mascara)

  cv2_imshow(img)
  cv2_imshow(destaque)



```

```python
analisar_plantacao("/content/plant.avif")
```
